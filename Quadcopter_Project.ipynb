## Controlling the Quadcopter

We provide a sample agent in the code cell below to show you how to use the sim to control the quadcopter.  This agent is even simpler than the sample agent that you'll examine (in `agents/policy_search.py`) later in this notebook!

The agent controls the quadcopter by setting the revolutions per second on each of its four rotors.  The provided agent in the `Basic_Agent` class below always selects a random action for each of the four rotors.  These four speeds are returned by the `act` method as a list of four floating-point numbers.  

For this project, the agent that you will implement in `agents/agent.py` will have a far more intelligent method for selecting actions!

```python
import random

class Basic_Agent():
    def __init__(self, task):
        self.task = task
    
    def act(self):
        new_thrust = random.gauss(450., 25.)
        return [new_thrust + random.gauss(0., 1.) for x in range(4)]
```

Run the code cell below to have the agent select actions to control the quadcopter.  

Feel free to change the provided values of `runtime`, `init_pose`, `init_velocities`, and `init_angle_velocities` below to change the starting conditions of the quadcopter.

The `labels` list below annotates statistics that are saved while running the simulation.  All of this information is saved in a text file `data.txt` and stored in the dictionary `results`.  

```python
%load_ext autoreload
%autoreload 2

import csv
import numpy as np
from task import Task

# Modify the values below to give the quadcopter a different starting position.
runtime = 5.                                     # time limit of the episode
init_pose = np.array([0., 0., 10., 0., 0., 0.])  # initial pose
init_velocities = np.array([0., 0., 0.])         # initial velocities
init_angle_velocities = np.array([0., 0., 0.])   # initial angle velocities
file_output = 'data.txt'                         # file name for saved results

# Setup
task = Task(init_pose, init_velocities, init_angle_velocities, runtime)
agent = Basic_Agent(task)
done = False
labels = ['time', 'x', 'y', 'z', 'phi', 'theta', 'psi', 'x_velocity',
          'y_velocity', 'z_velocity', 'phi_velocity', 'theta_velocity',
          'psi_velocity', 'rotor_speed1', 'rotor_speed2', 'rotor_speed3', 'rotor_speed4']
results = {x : [] for x in labels}

# Run the simulation, and save the results.
with open(file_output, 'w') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(labels)
    while True:
        rotor_speeds = agent.act()
        _, _, done = task.step(rotor_speeds)
        to_write = [task.sim.time] + list(task.sim.pose) + list(task.sim.v) + list(task.sim.angular_v) + list(rotor_speeds)
        for ii in range(len(labels)):
            results[labels[ii]].append(to_write[ii])
        writer.writerow(to_write)
        if done:
            break
```

## The Task

A sample task has been provided for you in `task.py`.  Open this file in a new window now. 

The `__init__()` method is used to initialize several variables that are needed to specify the task.  
- The simulator is initialized as an instance of the `PhysicsSim` class (from `physics_sim.py`).  
- Inspired by the methodology in the original DDPG paper, we make use of action repeats.  For each timestep of the agent, we step the simulation `action_repeats` timesteps.  If you are not familiar with action repeats, please read the **Results** section in [the DDPG paper](https://arxiv.org/abs/1509.02971).
- We set the number of elements in the state vector.  For the sample task, we only work with the 6-dimensional pose information.  To set the size of the state (`state_size`), we must take action repeats into account.  
- The environment will always have a 4-dimensional action space, with one entry for each rotor (`action_size=4`). You can set the minimum (`action_low`) and maximum (`action_high`) values of each entry here.
- The sample task in this provided file is for the agent to reach a target position.  We specify that target position as a variable.

The `reset()` method resets the simulator.  The agent should call this method every time the episode ends.  You can see an example of this in the code cell below.

The `step()` method is perhaps the most important.  It accepts the agent's choice of action `rotor_speeds`, which is used to prepare the next state to pass on to the agent.  Then, the reward is computed from `get_reward()`.  The episode is considered done if the time limit has been exceeded, or the quadcopter has travelled outside of the bounds of the simulation.

## The Agent

The sample agent given in `agents/policy_search.py` uses a very simplistic linear policy to directly compute the action vector as a dot product of the state vector and a matrix of weights. Then, it randomly perturbs the parameters by adding some Gaussian noise, to produce a different policy. Based on the average reward obtained in each episode (`score`), it keeps track of the best set of parameters found so far, how the score is changing, and accordingly tweaks a scaling factor to widen or tighten the noise.

Run the code cell below to see how the agent performs on the sample task.

```python
import sys
import pandas as pd
from agents.policy_search import PolicySearch_Agent
from task import Task

num_episodes = 1000
target_pos = np.array([0., 0., 10.])
task = Task(target_pos=target_pos)
agent = PolicySearch_Agent(task) 

for i_episode in range(1, num_episodes+1):
    state = agent.reset_episode() # start a new episode
    while True:
        action = agent.act(state) 
        next_state, reward, done = task.step(action)
        agent.step(reward, done)
        state = next_state
        if done:
            print("\rEpisode = {:4d}, score = {:7.3f} (best = {:7.3f}), noise_scale = {}".format(
                i_episode, agent.score, agent.best_score, agent.noise_scale), end="")  # [debug]
            break
    sys.stdout.flush()
```

```python
## TODO: Train your agent here.
import sys
import pandas as pd
import numpy as np
from agents.agent import DDPG
from task import Task

num_episodes = 2000
target_pos = np.array([0., 0., 50.])
task = Task(target_pos=target_pos)
agent = DDPG(task)

scores = [] # store rewards accumulated in episodes
best_score = -np.inf
for i_episode in range(1, num_episodes+1):
    state = agent.reset_episode() # start a new episode
    # var to keep track of total reward accumulated in an episode
    score = 0
    while True:
        action = agent.act(state) 
        next_state, reward, done = task.step(action)
        agent.step(action, reward, next_state, done)
        state = next_state
        score += reward 
        if done:
            scores.append(score)
            best_score = max(best_score, score)
            print("\rEpisode = {:4d}, score = {:7.3f} (best = {:7.3f})".format(i_episode,
                                                                               score, best_score), end="")
            break
    sys.stdout.flush()
```

```python
import csv

# Do a production run
labels = ['time', 'x', 'y', 'z', 'phi', 'theta', 'psi', 'x_velocity',
          'y_velocity', 'z_velocity', 'phi_velocity', 'theta_velocity',
          'psi_velocity', 'rotor_speed1', 'rotor_speed2', 'rotor_speed3', 'rotor_speed4']
results = {x : [] for x in labels}
file_output = "DDPG_agent.csv"

    
# Run one production episode and save the results.
rewards = []
with open(file_output, 'w') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(labels)
    state = agent.reset_episode()
    while True:
        rotor_speeds = agent.act(state, testing=True)
        next_state, reward, done = task.step(rotor_speeds)
        # write to file
        to_write = [task.sim.time] + list(task.sim.pose) + list(task.sim.v) + \
            list(task.sim.angular_v) + list(rotor_speeds)
        for ii in range(len(labels)):
            results[labels[ii]].append(to_write[ii])
        writer.writerow(to_write)
        # update observed state of the agent
        state = next_state
        if done:
            break
        else:
            rewards.append(reward)
```

```python
# Plot the test run results
def plot_episode_params(results):
    """plot a list of episode parameters as a func of time
    params - list of parameters to plot, taken from labels"""
    fig = plt.figure(figsize=(14, 12))
    grid = plt.GridSpec(3, 2)
    
    xyz_coord = ["x", "y", "z"]
    angle_coord = ["phi", "theta", "psi"]
    speed = ["x_velocity", "y_velocity", "z_velocity"]
    angular_speed = ["phi_velocity", "theta_velocity", "psi_velocity"]
    rotor_speeds = ["rotor_speed1", "rotor_speed2", "rotor_speed3", "rotor_speed4"]
    
    def plot_single(params, ax):
        "makes a single plot of a set of params wrt time"
        for p in params:
            ax.plot(results['time'], results[p], label=p)
        ax.legend()
        ax.set_xlabel("time")
    
    # plot states
    plot_single(xyz_coord, fig.add_subplot(grid[0,0]))
    plot_single(angle_coord, fig.add_subplot(grid[0,1]))
    plot_single(speed, fig.add_subplot(grid[1,0]))
    plot_single(angular_speed, fig.add_subplot(grid[1,1]))
    
    # plot rotor speeds
    plot_single(rotor_speeds, fig.add_subplot(grid[2,0:]))
    
    fig.show()
```

## Reflections

**Question 1**: Describe the task that you specified in `task.py`.  How did you design the reward function?

**Answer**: The task was to reach the coordinate (0, 0, 50) and hover there for the remainder of the episode (5 s). The reward function is designed as follows:
* First, a distance to the target position i calculated using the equation: $d=||\vec{x_{current}} - \vec{x_{target}}||$
* Then, the reward is calculated using the equation: $$reward = \frac{1}{1 + \alpha d}$$ where $\alpha = 0.01$ if $z \geq 50$ and $\alpha = 0.05$ otherwise. This scales the reward to a 0 - 1 range where 0 corresponds to infinite distance and 1 to 0 distance. This reward system encourages the quadcopter to reach the target position and stay there to collect as much reward as possible. The slower reward decay rate $\alpha$ when the quadcopter is above the target height is due to the experienced difficulty of getting the quadcopter to reach that height without dropping and crashing to the ground.
* Scaling the reward to a 0 - 1 range was hugely important due to issues in ANN training stability (as discussed bellow). Suggested in https://github.com/WittmannF/quadcopter-best-practices.
* Furthermore, the agent gets an additional reward of -10 when it crashes the quadcopter. This is checked by inspecting if the episode end time is less than the simulation runtime (taken from: https://github.com/WittmannF/quadcopter-best-practices)
* Due to the issues of agent training stability (the models were ANNs), I had to significantly reduce the action range that was available to the agent to only 850 - 860 rpm for each rotor (suggested in https://github.com/WittmannF/quadcopter-best-practices).

**Answer**: As a baseline, I took the suggested agent from the project guidelines. The agent needed to operate in continuous action space so the DDPG agent proposed seemed like the best idea. I first tested the agent on the selected task without any modifications, and then tried to improve it in various ways:
* The baseline agent kept crashing the quadcopter shortly after simulation start and didn't learn even after 2000 episodes.
* To add more information about the system that the agent could use to learn better policies, I added the quadcopter velocity and angular velocity to the available state information.
* Although this didn't significantly improve the performance, it caused the agent to learn slightly better policies.
* I speculated that the main issue with training this agent was lack of input normalization for the Critic and Actor neural nets. While it may be possible to normalize the x, y, z coordinates and the euler angles to a 0 - 1 range (we know what the min/max values are), this is not trivial for the velocity values. To mitigate the issue somewhat and speed up training, I added batch normalization layers after each hidden layer in the Critic and Actor models. This resulted is somewhat better performance.
* I then added an additional hidden layer (`Dense(units=32, activation="relu")`) in the Critic model right after the layer that combines the state and action pathways in order to "mix" the information in the two somewhat. To improve stability I also added regularization via dropout layers (`p=0.2`) after each hidden layer in both the Critic and Actor networks.
* I further experimented with the noise parameters of the OU process ($\sigma$) and the soft update parameter ($\tau$) for the target network update, but found the initial values to lead to better results.
